```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# open cv uses bgr colour code
img = cv2.imread("watch1.jpg")
cv2.rectangle(img, (70,70), (350,300), (0,255,0), 10)
#cv2.imshow("image", img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread("watch2.jpg")
cv2.circle(img, (100,100), 70, (255,0,0))
plt.imshow(img, cmap='gray', interpolation='bicubic')
#plt.imshow(img)
plt.show()
```

```python
img = cv2.imread("watch1.jpg")
points = np.array([[23,23], [167,178], [289,290], [200,69], [500,300]], np.int32)
cv2.polylines(img, [points], True, (0,255,0), 5)
plt.imshow(img)
plt.show()
```

```python
cv2.imread("watch1.jpg")
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, "Ashutosh!!", (50,300), font, 3, (0,255,0), 5, cv2.LINE_AA)
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
